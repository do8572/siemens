```python
def var():
    x = [1,3,5,10,15]
    rank_FM = np.load("./results/avg_same.npy", allow_pickle=True)
    rank_FM = rank_FM.tolist()
    rank_FM_new = np.load("./results/avg_diff.npy", allow_pickle=True)
    rank_FM_new = rank_FM_new.tolist()

    df = pd.read_csv("./results/dummy_hit_variations.csv")
    Deep_FM = df.iloc[np.array(range(1,7))*3-1,1:3]
    Deep_FM = Deep_FM.iloc[[0,2,3,4,5],:]
    df = pd.read_csv("./results/dummy_hit_variation_new.csv")
    Deep_FM_new = df.iloc[np.array(range(1,7))*3-1,1:3]
    Deep_FM_new = Deep_FM_new.iloc[[0,2,3,4,5],:]

    df = pd.read_csv("./results/variation.csv")
    baseline = df.iloc[:,1:]
    df = pd.read_csv("./results/variation_new.csv")
    baseline_new = df.iloc[:,1:]

    rank_FM_mean = pd.DataFrame(rank_FM).mean().values
    rank_FM_std = pd.DataFrame(rank_FM).std().values
    rank_FM_std_new = pd.DataFrame(rank_FM_new).std().values
    rank_FM_mean_new = pd.DataFrame(rank_FM_new).mean().values

    Deep_FM_mean = Deep_FM.iloc[:,0]
    Deep_FM_std = Deep_FM.iloc[:,1]
    Deep_FM_new_mean = Deep_FM_new.iloc[:,0]
    Deep_FM_new_std = Deep_FM_new.iloc[:,1]


    baseline_mean = baseline.iloc[:,0]
    baseline_std = baseline.iloc[:,1]
    baseline_new_mean = baseline_new.iloc[:,0]
    baseline_new_std = baseline_new.iloc[:,1]

    #baseline
    ax[0].plot(x, baseline_mean, 'g--', label="baseline any")
    ax[0].errorbar(x, baseline_mean, yerr=baseline_std/np.sqrt(6), fmt='.k')
    ax[0].plot(x, baseline_new_mean, 'g', label="baseline only new")
    ax[0].errorbar(x, baseline_new_mean, yerr=baseline_new_std/np.sqrt(6), fmt='.k')

    #rank_FM
    ax[0].plot(x, rank_FM_mean, 'r--', label="RankFM any")
    ax[0].errorbar(x, rank_FM_mean, yerr=rank_FM_std/np.sqrt(6), fmt='.k')
    ax[0].plot(x, rank_FM_mean_new, 'r',label="RankFM only new")
    ax[0].errorbar(x, rank_FM_mean_new, yerr=rank_FM_std_new/np.sqrt(6), fmt='.k')

    #Deep_FM
    ax[0].plot(x, Deep_FM_mean, 'b--', label="DeepFM any")
    ax[0].errorbar(x, Deep_FM_mean, yerr=Deep_FM_std/np.sqrt(6), fmt='.k')
    ax[0].plot(x, Deep_FM_new_mean, 'b', label="DeepFM only new")
    ax[0].errorbar(x, Deep_FM_new_mean, yerr=Deep_FM_new_std/np.sqrt(6), fmt='.k')

    ax[0].set_xticks(x)
    ax[0].set_ylim(0,1.1)
    ax[0].set_xlabel('k')
    ax[0].set_ylabel('hit_rate')
    ax[0].set_title("Variations", fontsize=15, bbox={'facecolor':'0.9', 'pad':2}, loc="left", x=0.055, y=0.865)
    
```

```python
def prod():
    rank_FM = np.load("./results/avg_same_prod.npy", allow_pickle=True)
    rank_FM = rank_FM.tolist()
    rank_FM_new = np.load("./results/avg_diff_prod.npy", allow_pickle=True)
    rank_FM_new = rank_FM_new.tolist()
    df = pd.read_csv("./results/dummy_hit_products.csv")
    Deep_FM = df.iloc[np.array(range(1,7))*3-1,1:3]
    Deep_FM = Deep_FM.iloc[[0,2,3,4,5],:]
    df = pd.read_csv("./results/dummy_hit_products_new.csv")
    Deep_FM_new = df.iloc[np.array([0,2,4,5,6]),1:3]
    df = pd.read_csv("./results/product.csv")
    baseline = df.iloc[:,1:]
    df = pd.read_csv("./results/product_new.csv")
    baseline_new = df.iloc[:,1:]
    x = [1,3,5,10,15]

    rank_FM_mean = pd.DataFrame(rank_FM).mean().values
    rank_FM_std = pd.DataFrame(rank_FM).std().values
    rank_FM_std_new = pd.DataFrame(rank_FM_new).std().values
    rank_FM_mean_new = pd.DataFrame(rank_FM_new).mean().values

    Deep_FM_mean = Deep_FM.iloc[:,0]
    Deep_FM_std = Deep_FM.iloc[:,1]
    Deep_FM_new_mean = Deep_FM_new.iloc[:,0]
    Deep_FM_new_std = Deep_FM_new.iloc[:,1]


    baseline_mean = baseline.iloc[:,0]
    baseline_std = baseline.iloc[:,1]
    baseline_new_mean = baseline_new.iloc[:,0]
    baseline_new_std = baseline_new.iloc[:,1]

    #rank_FM

    ax[1].plot(x, rank_FM_mean, 'r--')
    ax[1].errorbar(x, rank_FM_mean, yerr=rank_FM_std/np.sqrt(6), fmt='.k')
    ax[1].plot(x, rank_FM_mean_new, 'r')
    ax[1].errorbar(x, rank_FM_mean_new, yerr=rank_FM_std_new/np.sqrt(6), fmt='.k')

    #Deep_FM
    ax[1].plot(x, Deep_FM_mean, 'b--')
    ax[1].errorbar(x, Deep_FM_mean, yerr=Deep_FM_std/np.sqrt(6), fmt='.k')
    ax[1].plot(x, Deep_FM_new_mean, 'b')
    ax[1].errorbar(x, Deep_FM_new_mean, yerr=Deep_FM_new_std/np.sqrt(6), fmt='.k')

    #baseline
    ax[1].plot(x, baseline_mean, 'g--')
    ax[1].errorbar(x, baseline_mean, yerr=baseline_std/np.sqrt(6), fmt='.k')
    ax[1].plot(x, baseline_new_mean, 'g')
    ax[1].errorbar(x, baseline_new_mean, yerr=baseline_new_std/np.sqrt(6), fmt='.k')

    ax[1].set_xticks(x)
    ax[1].set_ylim(0,1.1)
    ax[1].set_xlabel('k')
    ax[1].set_ylabel('hit_rate')
    ax[1].set_title("Products", fontsize=15, bbox={'facecolor':'0.9', 'pad':2}, loc="left", x=0.055, y=0.865)
```

```python
def fam():
    rank_FM = np.load("./results/avg_same_fam.npy", allow_pickle=True)
    rank_FM = rank_FM.tolist()
    rank_FM_new = np.load("./results/avg_diff_fam.npy", allow_pickle=True)
    rank_FM_new = rank_FM_new.tolist()
    df = pd.read_csv("./results/dummy_hit_family.csv")
    Deep_FM = df.iloc[np.array(range(1,7))*3-1,1:3]
    Deep_FM = Deep_FM.iloc[[0,2,3,4,5],:]
    df = pd.read_csv("./results/deepfm_hit_family_new.csv")
    Deep_FM_new = df.iloc[np.array([0,2,4,5,6]),1:3]
    df = pd.read_csv("./results/family.csv")
    baseline = df.iloc[:,1:]
    df = pd.read_csv("./results/family_new.csv")
    baseline_new = df.iloc[:,1:]
    x = [1,3,5,10,15]

    rank_FM_mean = pd.DataFrame(rank_FM).mean().values
    rank_FM_std = pd.DataFrame(rank_FM).std().values
    rank_FM_std_new = pd.DataFrame(rank_FM_new).std().values
    rank_FM_mean_new = pd.DataFrame(rank_FM_new).mean().values

    Deep_FM_mean = Deep_FM.iloc[:,0]
    Deep_FM_std = Deep_FM.iloc[:,1]
    Deep_FM_new_mean = Deep_FM_new.iloc[:,0]
    Deep_FM_new_std = Deep_FM_new.iloc[:,1]


    baseline_mean = baseline.iloc[:,0]
    baseline_std = baseline.iloc[:,1]
    baseline_new_mean = baseline_new.iloc[:,0]
    baseline_new_std = baseline_new.iloc[:,1]

    #rank_FM

    ax[2].plot(x, rank_FM_mean, 'r--')
    ax[2].errorbar(x, rank_FM_mean, yerr=rank_FM_std/np.sqrt(6), fmt='.k')
    ax[2].plot(x, rank_FM_mean_new, 'r')
    ax[2].errorbar(x, rank_FM_mean_new, yerr=rank_FM_std_new/np.sqrt(6), fmt='.k')

    #Deep_FM
    ax[2].plot(x, Deep_FM_mean, 'b--')
    ax[2].errorbar(x, Deep_FM_mean, yerr=Deep_FM_std/np.sqrt(6), fmt='.k')
    ax[2].plot(x, Deep_FM_new_mean, 'b')
    ax[2].errorbar(x, Deep_FM_new_mean, yerr=Deep_FM_new_std/np.sqrt(6), fmt='.k')

    #baseline
    ax[2].plot(x, baseline_mean, 'g--')
    ax[2].errorbar(x, baseline_mean, yerr=baseline_std/np.sqrt(6), fmt='.k')
    ax[2].plot(x, baseline_new_mean, 'g')
    ax[2].errorbar(x, baseline_new_mean, yerr=baseline_new_std/np.sqrt(6), fmt='.k')

    ax[2].set_xticks(x)
    ax[2].set_ylim(0,1.1)
    ax[2].set_xlabel('k')
    ax[2].set_ylabel('hit_rate')
    ax[2].set_title("Families", fontsize=15, bbox={'facecolor':'0.9', 'pad':2}, x=0.8, y=0.1)
```
